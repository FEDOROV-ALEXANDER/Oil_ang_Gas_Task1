```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d

# Параметры сетки
length = 20  # Длина прямоугольника
width = 20   # Ширина прямоугольника
Nx = 10      # Количество точек по оси X
Ny = 10      # Количество точек по оси Y

# Создание прямоугольной сетки
X, Y = np.meshgrid(np.linspace(0, length, Nx), np.linspace(0, width, Ny))
points = np.vstack([X.ravel(), Y.ravel()]).T

# Центральная точка для уточнения сетки Вороного
center_point = np.array([10, 10])

# Добавление центральной точки к набору точек
points = np.vstack([points, center_point])

# Функция для уточнения сетки с несколькими слоями
def refine_voronoi_with_layers(points, center_point, length, width, refinement_factor=0.5, num_layers=3):
    # Вычисление диаграммы Вороного
    vor = Voronoi(points)

    # Уточнение сетки: добавление новых точек вблизи центральной точки
    new_points = []
    current_radius = 1.0
    for layer in range(num_layers):
        for angle in np.linspace(0, 2 * np.pi, 36, endpoint=False):
            new_points.append([center_point[0] + current_radius * np.cos(angle), center_point[1] + current_radius * np.sin(angle)])
        current_radius *= refinement_factor

    # Объединение исходных и новых точек
    all_points = np.vstack([points, new_points])

    # Вычисление новой диаграммы Вороного
    refined_vor = Voronoi(all_points)
    return refined_vor

# Уточнение сетки Вороного с несколькими слоями
refined_vor = refine_voronoi_with_layers(points, center_point, length, width)

# Получение центров ячеек Вороного
centers = refined_vor.points

# Визуализация уточненной сетки Вороного
voronoi_plot_2d(refined_vor)
plt.plot(center_point[0], center_point[1], 'ko')  # Отмечаем центральную точку
plt.plot(centers[:, 0], centers[:, 1], 'ro')  # Отмечаем центры ячеек красными кружками
plt.xlim(0, length)
plt.ylim(0, width)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

# Вывод центров ячеек
print("Центры ячеек Вороного:")
print(centers)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d

# Параметры сетки
length = 20  # Длина прямоугольника
width = 20   # Ширина прямоугольника
Nx = 10      # Количество точек по оси X
Ny = 10      # Количество точек по оси Y

# Создание прямоугольной сетки
X, Y = np.meshgrid(np.linspace(0, length, Nx), np.linspace(0, width, Ny))
points = np.vstack([X.ravel(), Y.ravel()]).T

# Центральная точка для уточнения сетки Вороного
center_point = np.array([10, 10])

# Добавление центральной точки к набору точек
points = np.vstack([points, center_point])

# Функция для уточнения сетки с круговой областью и переменным количеством точек
def refine_voronoi_with_circular_region(points, center_point, length, width, refinement_factor=0.5, num_layers=3):
    # Вычисление диаграммы Вороного
    vor = Voronoi(points)

    # Уточнение сетки: добавление новых точек в круговой области вокруг центральной точки
    new_points = []
    current_radius = 1.0
    for layer in range(num_layers):
        num_points_on_circle = 36 // (layer + 1)  # Уменьшаем количество точек на окружности с каждым слоем
        for angle in np.linspace(0, 2 * np.pi, num_points_on_circle, endpoint=False):
            new_points.append([center_point[0] + current_radius * np.cos(angle), center_point[1] + current_radius * np.sin(angle)])
        current_radius *= refinement_factor

    # Объединение исходных и новых точек
    all_points = np.vstack([points, new_points])

    # Вычисление новой диаграммы Вороного
    refined_vor = Voronoi(all_points)
    return refined_vor

# Уточнение сетки Вороного с круговой областью и переменным количеством точек
refined_vor = refine_voronoi_with_circular_region(points, center_point, length, width)

# Получение центров ячеек Вороного
centers = refined_vor.points

# Визуализация уточненной сетки Вороного
voronoi_plot_2d(refined_vor)
# plt.plot(center_point[0], center_point[1], 'ko')  # Отмечаем центральную точку
# plt.plot(centers[:, 0], centers[:, 1], 'ro')  # Отмечаем центры ячеек красными кружками
plt.xlim(0, length)
plt.ylim(0, width)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

# Вывод центров ячеек
print("Центры ячеек Вороного:")
print(centers)
```
